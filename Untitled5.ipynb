4. C помощью метода локтя определите оптимальное количество кластеров и разделите
данные на кластеры методом K-means

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = pd.read_csv('WineQT.csv')

features = [
    'fixed acidity',
    'volatile acidity',
    'alcohol',
    'pH'
]

X = data[features]

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

inertia = []
k_range = range(1, 11)  # Проверяем от 1 до 10 кластеров

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_normalized)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Метод локтя для определения оптимального количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')
plt.xticks(k_range)
plt.grid()
plt.show()

optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_normalized)

data['Cluster'] = clusters
print(data.head())
```

7. Визуализируйте результаты кластеризации методом иерархической кластеризации.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

data = pd.read_csv('WineQT.csv')

features = [
    'fixed acidity',
    'volatile acidity',
    'alcohol',
    'pH'
]

X = data[features]

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Проведение иерархической кластеризации
linked = linkage(X_normalized, method='ward')

# Определение количества кластеров
optimal_k = 3
clusters = fcluster(linked, optimal_k, criterion='maxclust')

data['Cluster'] = clusters

plt.figure(figsize=(10, 6))
plt.scatter(data['fixed acidity'], data['alcohol'], c=data['Cluster'], cmap='viridis', marker='o')
plt.title('Визуализация кластеров (иерархическая кластеризация)')
plt.xlabel('Фиксированная кислотность')
plt.ylabel('Уровень алкоголя')
plt.colorbar(label='Кластер')
plt.grid()
plt.show()
```

Оцените качество кластеризации методами K-means и иерархической кластеризации,
рассчитав пару метрик качества кластеризации (модуль sklearn.metrics). Например, силуэт
для выборки silhouette_score() и др.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.cluster.hierarchy import linkage, fcluster

data = pd.read_csv('WineQT.csv')

features = [
    'fixed acidity',
    'volatile acidity',
    'alcohol',
    'pH'
]

X = data[features]

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# K-means кластеризация
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_normalized)
kmeans_clusters = kmeans.labels_

# Иерархическая кластеризация
linked = linkage(X_normalized, method='ward')
hierarchical_clusters = fcluster(linked, 3, criterion='maxclust')

# Оценка качества кластеризации K-means
kmeans_silhouette = silhouette_score(X_normalized, kmeans_clusters)
kmeans_davies_bouldin = davies_bouldin_score(X_normalized, kmeans_clusters)

# Оценка качества кластеризации иерархической кластеризации
hierarchical_silhouette = silhouette_score(X_normalized, hierarchical_clusters)
hierarchical_davies_bouldin = davies_bouldin_score(X_normalized, hierarchical_clusters)

print("K-means:")
print(f"Silhouette Score: {kmeans_silhouette:.4f}")
print(f"Davies-Bouldin Index: {kmeans_davies_bouldin:.4f}")

print("\nИерархическая кластеризация:")
print(f"Silhouette Score: {hierarchical_silhouette:.4f}")
print(f"Davies-Bouldin Index: {hierarchical_davies_bouldin:.4f}")
```

Из датасета выберите любой конкретный
объект (если вы делаете модель на датасете
Country-data.csv, то выберите любую страну) и
визуализируйте этот объект в виде точки
отличного цвета и размера на графике
кластеров (пример на рисунке, точка
пурпурного цвета).

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, fcluster

data = pd.read_csv('WineQT.csv')

features = [
    'fixed acidity',
    'volatile acidity',
    'alcohol',
    'pH'
]

X = data[features]

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# K-means кластеризация
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_normalized)
data['Cluster'] = kmeans.labels_

# Выбор конкретного объекта (например, первого объекта в датасете)
specific_object_index = 0
specific_object = data.loc[specific_object_index]

# Визуализация кластеров
plt.figure(figsize=(10, 6))

# Отображение всех кластеров
for cluster in data['Cluster'].unique():
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['fixed acidity'], cluster_data['alcohol'], label=f'Cluster {cluster}')

# Выделение конкретного объекта
plt.scatter(specific_object['fixed acidity'], specific_object['alcohol'],
            color='purple', s=100, label='Specific Object', edgecolor='black')

plt.title('Визуализация кластеров с конкретным объектом')
plt.xlabel('Фиксированная кислотность')
plt.ylabel('Уровень алкоголя')
plt.legend()
plt.grid()
plt.show()
```
